# KNN

KNN (K-Nearest Neighbor) is a simple supervised classification algorithm we can use to assign a class to new data point. It can be used for regression as well, KNN does not make any assumptions on the data distribution, hence it is non-parametric. It keeps all the training data to make future predictions by computing the similarity between an input sample and each training instance.

1.   Compute the distance between the new data point with every training example.
2.   For computing the distance measures such as Euclidean distance, Hamming distance or Manhattan distance will be used.
3.   Model picks K entries in the training set, the closest to the new data point.
4.   Then it does the majority vote, i.e. the most common class/label among those K entries will be the class of the new data point.


# Excercise

### Let's find the best *k*, i.e. the value of neighbors that maximize the accuracy in classification

To start, I divide the dataset into ttaining data and testing data, so I can maximise the accuracy on both dsamples to have a model which isn't too overfitting or too underfitting.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

I define a function to get the mean score of the cross validation according to a specified k value

```python
def simulate(k):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(knn_model, X_train, y_train, cv=5)
    print(k,score.mean())
    return score.mean()
```

```python
scores = []
k_range = range(1,50)

for k in k_range:
    scores.append(simulate(k))
    

    
plt.plot(k_range, scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

From the graph we see that there are 2 values where the mean accuracy obtained using the cross validation is maximized, so we have to find them and test those value on the testing set to see which is better

```python
M = max(scores)
better_ks = []
for i in range(0,len(scores)):
    if scores[i] == M:
        better_ks.append(i+1)
        print(i+1,scores[i])
```

Now I want to train my dataset and get some predictions, according to different ks. Therefore I want to measure the accuracy over the testing set, to find the best parameter according to this view of the classification.

```python
accuracies = []
for k in k_range:
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train,y_train)
    y_pred = knn_model.predict(X_test)
    dat = classification_report(y_test, y_pred, target_names=iris.target_names,output_dict=True)["accuracy"]
    print(str(k),str(dat))
    accuracies.append(dat)
    
plt.plot(k_range, accuracies)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

So we see that both k = 7 and k = 12 are maximized in both training and testing sets, therefore they're both a optimal choice.
